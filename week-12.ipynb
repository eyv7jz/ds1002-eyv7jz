## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt
colors = ['blue', 'red', 'turquoise', 'green', 'orange']
plt.pie(slices, labels=labels, colors=colors, startangle=90, explode = (0, 0, 0, 0.1, 0), autopct = '%1.2f%%')
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)
```

```python
plt.hist(n, bins=100, color='m')
plt.title("Histogram")
plt.show()
```

## Scatterplot with Trendline (3 points)

```python
# Create a scatterplot with a trendline from two lists of values

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

# If you want them combined into a dataframe
df = pd.DataFrame({'x': x, 'y': y})
```

```python
import seaborn as sns

plt.scatter(x, y, color='c')
sns.regplot(x='x', y='y', data=df, scatter=False, color='black')
plt.title('Scatterplot')
plt.show()
```
